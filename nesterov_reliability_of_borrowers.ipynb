### Заполнение пропусков

На этапе обзора данных было замечено, что пропущены данные об общем трудовом стаже и ежемесячном доходе клиента. Проверим ещё раз.

```python
# получение количества пропущенных значений в столбцах df
df.isna().sum()
```

Оценим их долю от общего количества значений.

```python
# вывод доли пропущенных значений
missed_part_of_days_employed = df['days_employed'].isna().sum()/df['days_employed'].count()
missed_part_of_total_income = df['total_income'].isna().sum()/df['total_income'].count()
print(f'Доля пропущенных значений об общем трудовом стаже {missed_part_of_days_employed:.0%}')
print(f'Доля пропущенных значений о ежемесячном доходе {missed_part_of_total_income:.0%}')
```

При обзоре данных можно было заметить, что значения обоих столбцов отсутствуют только единовременно для одной строки. Убедимся, что данное условие выполняется всегда. Сравним количество одновременных пропусков и пропусков в одном из взятых столбцов.

```python
# вывод количества пропусков
number_of_double_missed_values = len(df[(df['days_employed'].isna()) & (df['total_income'].isna())])
number_of_days_employed_missed_values = df['days_employed'].isna().sum()
print(f'Количество одновременных пропусков: {number_of_double_missed_values}')
print(f'Количество пропусков одного столбца: {number_of_days_employed_missed_values}')
```

Количество одинаковое, что подтверждает, что предположение верно.

Далее заполним пропуски в столбце без каких-либо ещё артефактов данных — `total_income`. Новым будет значение, характерное для ежемесячного дохода клиента; выберем медианное, так как разброс зарплат может быть очень большим. Из-за этого при выборе, например, среднего значения оно может оказаться недостаточно репрезентативным, так как возможен сильный перекос в сторону сильно отличающегося значения. 

```python
# замена пропущенных значений в столбце медианным
df['total_income'] = df['total_income'].fillna(df['total_income'].median())

# проверка отсутствия пропущенных значений ежемесячного дохода
df['total_income'].isna().sum()
```

**Промежуточные выводы**

Обработав пропуски в столбцах, можно прийти к определённым выводам:
1. Доля пропущенных значений — 11% — была достаточно велика, в связи с чем в одном из столбцов они были заменены на медианное значение количественного параметра, позволяющее не сильно исказить общую достоверность данных, а также не упустить необходимые для дальнейшего исследования значения.
2. Учитывая, что количество пропущенных значений в каждом случае совпадает и пропуски в обоих столбцах появляются только одновременно, можно сделать предположение, что данные пропуски неслучайны. Причиной их появления могло быть, например, намеренное непредоставление соответствующих данных клиентами с низкой зарплатой и небольшим трудовым стажем из-за их опасения в отказе выдачи кредита на основании полученной информации. Также это могут быть люди, ни разу в своей жизни не работающиее по найму и не получающие лично какого-либо дохода, что тоже возможно.

### Проверка данных на аномалии и исправления

Приступим к обработке артефактов в данных перед избавлением от пропущенных значений в столбце `days_employed`.
Отрицательные значения и сотни тысяч дней трудового стажа не могут соответствовать действительности. В иделе можно было бы обратиться к тому, от кого получены данные за разъяснениями, но за неимением такой возможности произведем изменения соответствующих значений.

В отрицательных значениях, вероятно, ошибочно стоит знак "минус", так как модули значений выглядят вполне соответствующими реальности.
Переведём все значения трудового стажа в положительные.

```python
# замена значений в столбце их модулями
df['days_employed'] = abs(df['days_employed'])

# проверка отсутствия отрицательных значений столбца days_employed
df[df['days_employed'] < 0]['dob_years'].count()
```

При обзоре данных можно было заметить, что слишком большие значения трудового стажа встречаются в основном у пенсионеров. Для проверки предположения найдём средний показатель стажа для разных типов занятости клиентов.

```python
# нахождение всех значений столбца income_type
income_types = df['income_type'].unique()

# вывод на экран среднего значения для всех типов занятости
for type in income_types:
    median_days_employed = df[df['income_type']==type]['days_employed'].median()
    print(f'Среднее количество дней трудового стажа типа занятости "{type}": {median_days_employed}')
```

Нереалистично высокие показатели встречаются у двух типов клиентов: пенсионеров и безработных. Очевидные причины появления артефактов с такими значениями без дополнительной информации назвать нельзя. Можно предположить, что значения для данных категорий хранятся в другой еденице измерения, отличной от дней. По величине значений самой логичной выглядят часы.
Разделим значения в соответствующих колонках на 24 и выведем их новые средние показатели.

```python
# изменение значений в столбце income_type по нужному условию
income_type = df['income_type']
for index in range(len(income_type)):
    if df.loc[index, 'income_type'] == 'пенсионер' or df.loc[index, 'income_type'] == 'безработный':
        df.loc[index, 'days_employed'] /= 24
        
# вывод на экран обновленного среднего значения для всех типов занятости
for type in income_types:
    median_days_employed = df[df['income_type']==type]['days_employed'].median()
    print(f'Среднее количество дней трудового стажа типа занятости "{type}": {median_days_employed}')
```

Теперь значения выглядят более соответствующими действительности. Можем приступить к заполнению пропусков в столбце с трудовым стажем клентов.

```python
# замена пропущенных значений в столбце медианным
df['days_employed'] = df['days_employed'].fillna(df['days_employed'].median())

# проверка отсутствия пропущенных значений во всём датасете
df.isna().sum()
```

**Промежуточные итоги**

Так как в данных были обнаружены артефакты разного характера, они были обработаны для заполнения оставшихся пропусков датасета более релевантными значениями.
Найти точную причину появления подобных аномалий за неимением возможности обратиться к источнику информации не возможно. Остаётся предположить, что могла произойти техническая ошибка выгрузки или просчета значений.

Приведём все к одному типу, величина значений во всех столбцах позволяет это сделать.

```python
# изменение типа данных в нужных столбцах
df[['children', 'dob_years', 'education_id', 'family_status_id', 'debt']] = df[['children', 'dob_years', 'education_id', 'family_status_id', 'debt']].astype('int')

# проверка изменений
df.dtypes
```

```python
# получение случайных 15 строк таблицы df
df.sample(15)
```

### Удаление дубликатов

Одной из распространенных проблем с данными также является наличие дубликатов. При выводе случайных строк таблицы можно было заметить как минимум неявные дубликаты в столбце `education`, одинаковые значения, но записанные с использованием либо заглавных, либо строчных букв.

Для нахождения как можно большего количества строк с дубликатами изначально избавимся от этих и других возможных идентичных значений, записанных в разном регистре.

```python
# Изменение регистра в нужных столбцах
df[['education',
    'family_status',
    'income_type',
    'gender',
    'purpose']] = df[['education',
                      'family_status',
                      'income_type',
                      'gender',
                      'purpose']].apply(lambda x:x.str.lower(), axis=1)
    
# оценивание внесённых изменений
df.sample(15)
```

Далее проверим, присутствуют ли в данных явные строки-дубликаты и при наличии удалим их.

```python
# получение количества явных дубликатов
df.duplicated().sum()
```

```python
# удаление явных дубликатов
df = df.drop_duplicates().reset_index(drop=True)
```

В итоге неявные дубликаты были обнаржуены только в столбце с информацией о цели получения кредита. Обработаем их, заменив разные значения одним обобщающим, где это будет корректно по смыслу, то есть не нарушится детализация цели, так как последнее может понадобиться при какой-либо иной работе с представленными данными.
Автоматизируем этот процесс, использовав собственную функцию.

```python
# создание функции для замены неявных дубликатов
def replace_wrong_values(wrong_values, correct_value):
    for wrong_value in wrong_values:
        df['purpose'] = df['purpose'].replace(wrong_value, correct_value)
```

```python
# последовательный вызов функции или метода replace() для изменения соответствующих значений
duplicates = ['свадьба', 'на проведение свадьбы', 'сыграть свадьбу']
generalizing_value = 'проведение свадьбы'
replace_wrong_values(duplicates, generalizing_value)

duplicates = ['автомобиль', 'сделка с автомобилем']
generalizing_value = 'операции с автомобилем'
replace_wrong_values(duplicates, generalizing_value)

duplicates = ['образование', 'заняться образованием']
generalizing_value = 'получение образования'
replace_wrong_values(duplicates, generalizing_value)

duplicates = ['заняться высшим образованием', 'высшее образование']
generalizing_value = 'получение высшего образования'
replace_wrong_values(duplicates, generalizing_value)

df['purpose'] = df['purpose'].replace('покупка жилья', 'покупка жилой недвижимости')
df['purpose'] = df['purpose'].replace('недвижимость', 'операции с недвижимостью')
df['purpose'] = df['purpose'].replace('жилье', 'операции с жильем')
df['purpose'] = df['purpose'].replace('на покупку автомобиля', 'приобретение автомобиля')
df['purpose'] = df['purpose'].replace('дополнительное образование', 'получение дополнительного образования')
```

Посмотрим на изменения.

```python
# вывод уникальных значений столбца purpose
df['purpose'].value_counts()
```

**Промежуточные итоги**

1. При обработке дубликатов изначально был выполнен перевод всех соответствующих значений в один регистр для увеличеня количества явных дубликатов, так как избавляться от них проще. Неявные дубликаты были обнаружены методом `value_counts()` и заменены с использованием собственной функции и метода `replace()`.
2. Причиной появления явных дубликатов мог быть технический сбой при обработке данных. Разница же в регистре и дублирование одинаковых по смыслу целей получения кредита, по всей видимости, образовались из-за самостоятельного заполнения соответствующих данных разными людьми. Избежать этого можно было бы, например, дав клинтам при заполнении определенные категории на выбор.

### Категоризация дохода

Конечной целью исследования является определение того, от каких характеристик и особенностей клиентов зависит своевременный возврат ими кредита.
Единичные значения уровня дохода клиента и цели получения кредита достаточно имеют достаточно сильный разброс. Для получения релевантных статистических выводов на основе подобных данных к ним требуется применить категоризацию.

Начнем с категоризации значений дохода клиента.
Определимся с диапазонами категоризации. Для этого сформируем квантили столбца `total_income`.

```python
# получение квантилей значений дохода клиента
q = [0, 0.25, 0.5, 0.75, 1]
df_log['total_income'].quantile(q=q)
```

На основании значений разброса дохода сформируем столбец `total_income_category` со следующими категориями:

* 0–30000 — `E`
* 30001–50000 — `D`
* 50001–200000 — `C`
* 200001–1000000 — `B`
* 1000001 и выше — `A`

```python
# создание функции для категоризации значений
def total_income_category(income):
    if income <= 30000:
        return 'E'
    if income <= 50000:
        return 'D'
    if income <= 200000:
        return 'C'
    if income <= 1000000:
        return 'B'
    return 'A'

# тестирование функции
display(total_income_category(20000))
display(total_income_category(40000))
display(total_income_category(100000))
display(total_income_category(500000))
display(total_income_category(2000000))
```

Функция работает корректно, можно создавать столбец.

```python
# формирование столбца
df_log['total_income_category'] = df_log['total_income'].apply(total_income_category)

# оценивание внесённых изменений
df_log.sample(15)
```

### Категоризация целей кредита

Примением категоризацию и для столбца с целями порлучения кредита.
Ещё раз взглянем на уникальные значения `purpose`, чтобы определиться с соответствующими категориями.

```python
# вывод уникальных значений столбца purpose
df_log['purpose'].value_counts()
```

Создадим столбец `purpose_category`, в который войдут следующие 4 категории с соответствующими значениями:

* `проведение свадьбы`
* `получение образования`
* `операции с автомобилем`
* `операции с недвижимостью`

```python
# создание функции для категоризации значений
def purpose_category(purpose):
    if 'свадьб' in purpose:
        return 'проведение свадьбы'
    if 'образован' in purpose:
        return 'получение образования'
    if 'автомобил' in purpose:
        return 'операции с автомобилем'
    return 'операции с недвижимостью'

# тестирование функции
display(purpose_category('свадьба'))
display(purpose_category('образование'))
display(purpose_category('автомобиль'))
display(purpose_category('жилье'))
```

Всё в порядке, создаём новый столбец.

```python
# формирование столбца
df_log['purpose_category'] = df_log['purpose'].apply(purpose_category)

# оценивание внесённых изменений
df_log.sample(15)
```

### Выяснение зависимости между количеством детей клиента и возвратом им кредита в срок

Для начала взглянем на то, сколько детей в принципе бывает у клиентов из данного датасета.

```python
# вывод уникальных значений столбца
df_log['children'].value_counts()
```

Можно заметить, что присутствуют некорректные значения — "-1" ребёнок. Предположим, что это артефакт того же рода, что мы уже встречали ранее относительно столбца `days_employed`. Аналогично заменим значения на положительные.

```python
# замена значений в столбце их модулями
df_log['children'] = abs(df_log['children'])

# проверка отсутствия отрицательных значений столбца days_employed
df_log[df_log['children'] < 0]['dob_years'].count()
```

Построим сводную таблицу для оценивая того, насколько количество детей влияет на отсутствие просрочек оплаты кредита. В таблице присутстует показатель `debt`, значения которого говорят о том, была или нет просрочка. Оценим процентное соотношение вовремя закрытых кредитов ко всем кредитам в целом при разной численности детей в семье.
Так как проведение подобной операции понадобится нам и в дальнейшем для других характеристик клиентов, напишем собственную функцию, выводящую сводную таблицу с сортировкой по соотношению. 

```python
# создание функции
def relation(parametr):
    df_log_pivot = df_log.pivot_table(index=parametr, columns='debt', values='dob_years', aggfunc='count')
    df_log_pivot = df_log_pivot.fillna(0)
    df_log_pivot['percentage'] = df_log_pivot[0] / (df_log_pivot[0] + df_log_pivot[1]) * 100
    return df_log_pivot.sort_values(by='percentage')
```

```python
# вызов функии для оценки зависимости
relation('children')
```

Основываясь на полученных данных, можно сказать, что наблюдать какую-либо сильную зависимость между количеством детей и своевременным возвратом кредита нельзя.
Процент вовремя погашенного долга практически не отличается при изменении численности детей. Единственное, можно, конечно, отметить, что шанс своевременного возврата долга клиентом ниже всего при очень большом количестве детей, а выше — при их полном отсутвии. Но опять таки, между крайними значениями разница составляет всего 3%.
