# Исследование российского кинопроката

**Цель исследования** — выявить тенденции и особенности рынка российского кинопроката, при этом отдельно рассмотрев фильмы, получившие господдержку. Необходимо выяснить, насколько в подобных проектах заинтересован зритель.

Используемые данные опубликованы на портале открытых данных Министерства культуры Российской Федерации. Они содержат информацию о прокатных удостоверениях, сборах и государственной поддержке фильмов, а также информацию с сайта КиноПоиск.

**Ход исследования**

Откроем файлы с данными `data_movies.csv` и `data_shows.csv`, объединим их в один датафрейм и изучим его.

Проверим данные на пропуски, дубликаты и иные артефакты, исправим обнаруженные проблемы. Далее создадим необходимые дополнительные столбцы, после чего проведём исследовательский анализ данных.

Под конец же исследуем фильмы, получившие государственную поддержку и сформулируем общий вывод.
 
Таким образом, исследование пройдёт в пять этапов:
 1. Обзор данных.
 2. Предобработка данных.
 3. Анализ данных.
 4. Изучение фильмов с господдержкой.
 5. Общий вывод.

## Обзор данных

Изначально данные представлены в двух файлах. Для удобства дальнейшей работы объеденим их в один датафрейм. Сначала взглянем на основные особенности датасетов, чтобы понять, по какому общему признаку их можно объединить.

```python
# импорт библиотек и настройка отображения данных
import pandas as pd
import plotly.express as px
import warnings

pd.options.display.float_format = '{:.1f}'.format
```

```python
# чтение csv-файлов и создание датафреймов
movies = pd.read_csv('data_movies.csv')
shows = pd.read_csv('data_shows.csv')
```

```python
# получение первых строк таблиц
display(movies.head(10))
display(shows.head(10).style.format('{:.0f}'))
```

```python
# получение общей информации о данных в таблицах
print(movies.info())
print(shows.info())
```

Объединим датафреймы по совпадающему параметру — номеру прокатного удостоверения из столбцов `puNumber`.

В разных источниках данные значения записаны с использованием разных типов данных, поэтому сначала приведём нужные столбцы к одному типу. Перед этим убедимся, что все значения `puNumber` числовые.

После чего взглянем уже на общую информацию и распределнение значений итогового датафрейма.

```python
# поиск нечисловых значений puNumber
movies.query('not puNumber.str.isnumeric()').puNumber
```

```python
# удаление лишних пробелов
movies.puNumber = movies.puNumber.str.strip()

# проверка изменений
movies.query('not puNumber.str.isnumeric()').puNumber
```

Удалим данные о фильме без прокатного удостоверения.

```python
# добавление к первому датафрейму данных из второго после удаления строки и изменения типа данных
movies = movies.query('puNumber !="нет"').reset_index(drop=True)
movies.puNumber = movies.puNumber.astype('int')
movies = movies.merge(shows, on='puNumber', how='left')

# получение общей информации о данных датасета
movies.info()
```

```python
# получение данных о характерных значениях movies
movies.describe().style.format('{:.0f}')
```

3. Название столбца `puNumber` сформировано с использованием транслитерации и разного регистра; строковые значения столбцов записаны также в разных регистрах, что может затруднить обработку дубликатов.

## Предобработка данных

### Изменение типов данных

Изменим типы данных на более соответствующие характеру значений.

```python
# изменение числовых типов данных
movies[[
'refundable_support',
'nonrefundable_support'
]] = (
movies[[
'refundable_support',
'nonrefundable_support']]
.convert_dtypes()
.astype('Int64')
)
```

Для значений даты нам понадобится только день, округлим их соответствующим образом.

```python
# изменение типа данных для даты
movies.show_start_date = (
pd.to_datetime(
movies.show_start_date,
format='%Y-%m-%dT%H:%M:%S.000Z')
.dt.round('1D')
)
```

Перед изменением значений рейтинга взглянем на то, как они в принципе представлены.

```python
# получение значений столбца ratings
movies.ratings.unique()
```

Видно, что значения записаны в двух системах: дробной десятибальной и процентной. Приведём все значения столбца к одному представлению. 

```python
# удаление из значений символа процента и приведение к нужному типу
movies.ratings = movies.ratings.replace(
'%', '',
regex=True).astype('float')

# приведение всех значений к одному порядку
movies.loc[movies.ratings > 10,
'ratings'] = movies.ratings / 10
```

Взглянем на результаты изменений.

```python
# получение общей информации о данных датафрейма
movies.info()
```

```python
# получение случайных 10 строк таблицы movies
movies.sample(10, random_state=1)
```

### Исправление названия столбца и приведение значений к одному регистру

Исправим некорректное название одного столбца, а также приведём все значения датафрейма к одному регистру для более точных выполнений запросов и обработки дубликатов.

```python
# смена названия столбца
movies = movies.rename(columns={'puNumber': 'distribution_certificate_number'})

# приведение всех значений к нижнему регистру
for column in movies.columns[movies.dtypes=='object']:
    movies[column] = movies[column].str.lower()
```

Проверим успешность изменений.

```python
# получение случайных 10 строк таблицы movies
movies.sample(10, random_state=1)
```

### Обработка пропусков

В большинстве столбцов датафрема присутствуют пропуски. Рассмотрим их подробнее, чтобы определить, что можно сделать с пропущенными значениями в каждом случае.

Видно, что в столбцах `film_studio`, `production_country` и `director` доля пропусков мала. Проверим это и в случае подтверждения удалим строки с пропущенными значениями в данных столбцах.

```python
# получение доли пропусков в нужных столбцах
len(movies.query('film_studio.isna() or production_country.isna() or director.isna()')) / len(movies)
```

```python
# удаление строк с пропусками
movies = movies.dropna(
subset=[
'film_studio',
'production_country',
'director']).reset_index(drop=True)
```

Аналогично проверим долю пропущенных значений о продюссере и жанрах фильма.

```python
# получение доли пропусков в нужных столбцах
len(movies.query('producer.isna() or genres.isna()')) / len(movies)
```

Она составляет уже практически 20%, при удалении потеряется достаточно большая часть всего датасета. Заменить пропуски нечем, поэтому в данном случае оставим их как есть.

Наличие пропущенных значений с информацией о господдержке фильма обусловлено отсутствием оной для данных фильмов; пропуски для числовых значений в этих случаях можно заменить на 0. Перед заменой проверим, нет ли в таблице случаев, когда объём возвратных и невозвратных средств господдержки одновременно равен нулю. Это было бы неверно, так как изначально данные в соответствующих столбцах указаны только для фильмов, получивших государственную поддержку, соответственно, суммарно она не может быть нулевой.

Источник данных об общем бюджете фильма тот же, что и о господдержке, но заменить пропуски в данном случае каким-либо характерным значением, учитывая их высокую долю, будет неправильно. Также оставим их в изначальном виде. Похожая ситуация и с пропущенными значениями сборов, информация была менее, чем о половине фильмов, релевантые характерные значения получить невозможно.

```python
# проверка значений столбцов с данными о господдержке
len(movies.query('refundable_support==0 and nonrefundable_support==0'))
```

```python
# замена пропусков нулевым значением
movies[[
'refundable_support',
'nonrefundable_support'
]] = movies[[
'refundable_support',
'nonrefundable_support']].fillna(0)
```

Пропуски значений рейтинга фильмов при их не сильно большой доле можно заменить медианным значением. Проверим распредление этих пропусков по значениям некоторых столбцов, возможно, выявится какая-то закономерность, на основе которой можно будет сделать замену пропусков в данном случае более точной.

```python
# получение распределения пропусков по значениям
names = [
'type',
'production_country',
'age_restriction',
'financing_source',
'genres'
]
for name in names:
    display(movies.query('ratings.isna()')[name].value_counts().to_frame())
```

Большинство пропущенных значений рейтинга присутствуют у фильмов из России. Следовательно, замену будет логично производить, учитывая страну производства фильма.

```python
# замена пропущенных значений в столбце медианным c учётом страны производства
warnings.filterwarnings(action='error', message='Mean of empty slice')
for definition in movies.production_country.unique():
    try:
        movies.loc[
        movies.production_country==definition, 'ratings'
        ] = movies.loc[
        movies.production_country==definition, 'ratings'
        ].fillna(movies.loc[
        movies.production_country==definition, 'ratings'
        ].median().round(1))
    except:
        movies.loc[
        movies.production_country==definition, 'ratings'
        ] = movies.loc[
        movies.production_country==definition, 'ratings'
        ].fillna(movies.ratings.median().round(1))
                           
# проверка количества пропущенных значений во всём датасете
movies.isna().sum()
```

### Исправление категориальных значений

Поищем артефакты в категориальных параметрах датафрейма. Для этого для начала взглянем на их наиболее часто встречающиеся значения соответствующих столбцов.

Видим, что произошло неявное дублирование значений из-за лишнего пробела. Исправим это и аналогично обработаем остальные столбцы при необходимости.

```python
# удаление лишнего пробела в значениях
movies.type = movies.type.str.strip()

# проверка изменений
movies.type.value_counts()
```

```python
# получение значений столбца
movies.film_studio.value_counts().head(30)
```

```python
# получение значений столбца
movies.production_country.value_counts().head(60)
```

Можно заметить разные разделители, установим один общий для всех столбцов.

```python
# изменение значений
movies.production_country = (
movies.production_country.str.strip()
)
movies.production_country = (
movies.production_country
.replace(',', ', ', regex=True)
)
movies.production_country = (
movies.production_country
.replace(' - ', ', ', regex=True)
)
movies.production_country = (
movies.production_country
.replace(' -', ', ', regex=True)
)
movies.production_country = (
movies.production_country
.replace('- ', ', ', regex=True)
)
movies.production_country = (
movies.production_country
.replace('-', ', ', regex=True)
)

# проверка изменений
movies.production_country.value_counts().head(60)
```

```python
# получение значений столбца
movies.director.value_counts().head(30)
```

```python
# удаление лишнего пробела в значениях
movies.director = movies.director.str.strip()

# проверка изменений
movies.director.value_counts().head(30)
```

```python
# получение значений столбца
movies.producer.value_counts().head(30)
```

```python
# удаление лишнего пробела в значениях
movies.producer = movies.producer.str.strip()
```

```python
# получение значений столбца
movies.age_restriction.value_counts()
```

Данные категориальные значения столбца `age_restriction` лучше было бы записать с использованием просто цифр, так как по названию столбца будет понятно их значение, а анализировать удобнее, проще и эффективнее данные в числовом формате. Исправим значения соответствующим образом.

```python
# изменение значений
movies.age_restriction = (
movies.age_restriction
.str.extract('(\d+)',
expand=False).astype('int')
)

# проверка изменений
movies.age_restriction.value_counts()
```

### Особенности предоставления госсредств 

Проверим, были ли повторные прокаты одних и тех же фильмов с господдержкой.

```python
# обновление среза данных с господдержкой
gov_supported_films = movies.query('financing_source.notna()')

# получение дубликатов столбца title для фильмов с господдержкой
movies.query('financing_source.notna()').title.duplicated().sum()
```

Можно сказать, что в дальнейшем мы будем рассматривать не только отдельные прокаты, но и, собственно, фильмы.

```python
# объявление функции для создания таблицы и графика
def gov_supported_films_stat(parametr):
    stat = gov_supported_films.pivot_table(
        index=parametr,
        values=[
        'refundable_support',
        'nonrefundable_support'
        ], aggfunc=['sum','count'])
    stat['count', 'nonrefundable_support'] = (
        gov_supported_films.pivot_table(
        index=parametr,
        values='gov_support_part')
        ['gov_support_part']
    )
    stat.columns=[
        'Сумма невозвратных средств',
        'Сумма возвратных средств',
        'Доля господдержки',
        'Количество фильмов'
    ]
    stat.insert(
        2,
        'Итого выделено средств',
        stat.loc[
        :, 'Сумма невозвратных средств':'Сумма возвратных средств'
        ].sum(axis=1).astype('int64'))
    stat.insert(
        3,
        'Средняя сумма господдержки',
        (stat['Итого выделено средств']/
        stat['Количество фильмов']).astype('int64'))
    display(stat.sort_values(
        by='Итого выделено средств',
        ascending=False).style.background_gradient(cmap='Blues', axis=0))
    fig = px.bar(
        stat,
        y='Средняя сумма господдержки',
        width=1000, height=500,
        title='Объёмы господдержки').update_layout(
        yaxis_title='Средняя сумма выделеных средств',
        showlegend=False)
    return fig
```

```python
# вызов функции для изучения параметра
gov_supported_films_stat('show_start_year')
```

Общие объемы господдержки неравномерны от года к году. Больше всего средств от государства было выделено в 2019 году, притом что в нём же зафиксирована одна из самых низких долей госсредств от общего бюджета — в районе 50%. Наиболее низкие общие показатели видны в первые годы выделения господдержки — 2013, 2014 — в которые очень мало фильмов получили соответствующее финансирование.

Наибольший показатель господдержки, выделенной в среднем на фильм, видно в 2017 году, периоде также с одной из максимальных долей самостоятельного финансирования фильмов. Подобная тенденция может говорить как об увеличенной мотивации иных источников финансирования вкладываться в производство кино при значительных объемах господдержки, так и о готовности государства вкладывать достаточные средства при большей финансовой самостоятельности инициатора съёмок.

### Окупаемость фильмов с государственной поддержкой

Изучим, насколько хорошо окупаются фильмы, получившие поддержку от государства. Расмотрим общую окупаемость и среднюю окупаемость фильма по различным группам.

```python
# создание столбца со значениями окупаемости
gov_supported_films = gov_supported_films.assign(
payback = gov_supported_films.box_office /
gov_supported_films.budget)

# объявление функции для создания таблицы и графика
def gov_supported_films_payback(parametr):
    payback_stat = (
        gov_supported_films.query('box_office.notna()')
        .pivot_table(
        index=parametr,
        values='payback',
        aggfunc=['median', 'mean', 'count'])
    )
    payback_stat[['box_office', 'budget']] = (
        gov_supported_films.pivot_table(
        index=parametr,
        values=['box_office', 'budget'],
        aggfunc='sum')
        [['box_office', 'budget']]
    )
    payback_stat['payback'] = (
        payback_stat['box_office'] /
        payback_stat['budget']
    )
    payback_stat.columns = [
        'Медианная окупаемость',
        'Средняя окупаемость',
        'Количество фильмов',
        'Сборы',
        'Общий бюджет',
        'Общая окупаемость'
    ]
    display(payback_stat.sort_values(
        by='Медианная окупаемость',
        ascending=False)
        .style.format('{:.2f}')
        .background_gradient(cmap='Blues', axis=0))
    fig = px.bar(
        payback_stat,
        y='Медианная окупаемость',
        width=1000,
        height=500,
        title='Окупаемость фильмов').update_layout(
        yaxis_title='Средняя окупаемость фильма',
        showlegend=False)
    return fig
```

```python
# вызов функции для изучения параметра
gov_supported_films_payback('show_start_year')
```

При анализе правильнее будет ориентироваться на медианную окупаемость, более устойчивую к выбросам показателя сборов.

Можно видеть, что в среднем фильмы с господдержкой по отдельности в принципе не окупаются. Самое большое значение в районе 50% приходится на 2017 год. Очень низкий показатель 2013 года обусловлен в первую очередь всего двумя фильмами в прокате за этот год. Итогово только два года с самыми высокими сборами в целом — 2017 и 2018 — в сумме окупились положительно, на 154% и 142% соответственно.

```python
# вызов функции для изучения параметра
gov_supported_films_payback('type')
```

В разрезе по типам фильмов ни один полностью не окупился даже в целом за всё время. Медианная же окупаемость художественных фильмов вообще составляет очень низкие 14%, причём в прокате их было на порядок больше анимационных.

Медианная же окупаемость фильмов с определённой категорией не превышает 36%, а у фильмов с ограничением `18+`, имеющих в целом низкие сборы, но относительно высокий суммарный бюджет, и вовсе составляет всего 4%.

```python
# вызов функции для изучения параметра
gov_supported_films_payback('financing_source')
```

Положительно окупаются за всё время в целом (на 122%) только фильмы, финансируемые "Фондом кино", причем их приблизительно столько же, что и фильмов с поддержкой от Министерства культуры. Последние же и фильмы, совместо финансируемые обеими структурами, окупаются очень слабо. Их медианные показатели на фильм составляют 3-7%.
